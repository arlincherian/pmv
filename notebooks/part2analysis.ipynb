```python
import pandas as pd
import numpy as np
import altair as alt
from altair_saver import save
# alt.renderers.enable('altair_saver', fmts=['vega-lite', 'png'])
```

```python
# load pre survey data for 2021 
df = pd.read_excel("../data/raw-data/comp_data_part2.xlsx")
df.info()
```

# The Workshop content sustained my interest

```python
bar = alt.Chart(df, title="Workshop content interest of participants"
               ).transform_joinaggregate(
    total='count()',
    groupby=['workshop_interest']
).transform_calculate(
    percent=alt.datum.total/98
).mark_bar().encode(
    x= alt.X('percent:Q', title = "Response percerntage (%)", axis=alt.Axis(format='%')),
    y=alt.Y('workshop_interest', sort='x', title="")
    ).properties(
    width=400,
    height=150
    )

text = alt.Chart(df, title="Workshop content interest of participants"
               ).transform_joinaggregate(
    total='count()',
    groupby=['workshop_interest']
).transform_calculate(
    percent=alt.datum.total/98
).mark_text(
        align='left',
        baseline='middle',
        dx=3
).encode(
        y = alt.Y('workshop_interest', sort = 'x'),
        x = alt.X('percent:Q'),
        text=alt.Text('percent:Q',format='.1%')
).properties(
    width=800,
    height=250
    )

q1_plot = alt.layer(bar, text, data=df)

# Show the plot|
q1_plot
# save(q1_plot, 'q1_plot.png', scale_factor=2.0)
```
